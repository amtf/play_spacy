```python
stock_codes = df['Stock Code (LSE)'].drop_duplicates().tolist()
stock_codes[:5]
```

```python
company_names = df['Stock Name'].drop_duplicates().tolist()
company_names[:5]
```

```python
#現在由一個預設都不要, 從零做起
nlp = spacy.blank("en")
ruler = nlp.add_pipe('entity_ruler')

#險查只有entity_ruler
nlp.pipe_names
```

```python
patterns = []
for stock_code in stock_codes:
  patterns.append({'label': "STOCK", "pattern": stock_code})
for company_name in company_names:
  patterns.append({'label': "COMPANY", "pattern": company_name})
ruler.add_patterns(patterns)
```
